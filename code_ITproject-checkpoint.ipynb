```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# 1. WCZYTANIE DANYCH
file_name = 'global_data.csv'  
df = pd.read_csv(file_name)

# LISTA KOLUMN 
cols = [
    'GDP per Capita', 
    'Life Expectancy', 
    'Under 5 Mortality Rate',
    'Diet Calories Animal Protein' 
]

# ZMIANA NA LICZBY
for col in cols:
    
    df[col] = pd.to_numeric(df[col], errors='coerce')


df_clean = df.dropna(subset=cols).copy()

# ZMIANA NAZW
df_clean = df_clean.rename(columns={
    'GDP per Capita': 'GDP',
    'Under 5 Mortality Rate': 'Infant Mortality',
    'Diet Calories Animal Protein': 'Nutrition (Protein)'
})

print("DANE GOTOWE! GenerujÄ™ 4 wykresy w Twoich kolorach...")

# 2. KOLORY
MINT = "#80CBC4"    
PURPLE = "#7E57C2"  
DARK = "#002147"    
sns.set_style("whitegrid") 

# WYKRES 1: MACIERZ KORELACJI 
plt.figure(figsize=(10, 8))

numeric_df = df_clean[['GDP', 'Life Expectancy', 'Infant Mortality', 'Nutrition (Protein)']]

sns.heatmap(numeric_df.corr(), annot=True, cmap='PuBuGn', fmt=".2f", linewidths=1, linecolor='white')
plt.title('Correlation Matrix: Economy, Health & Nutrition', fontsize=14, color=DARK, fontweight='bold')
plt.savefig('chart_1_matrix.png', dpi=300, bbox_inches='tight')
plt.show()

#  WYKRES 2: GDP vs LIFE EXPECTANCY 
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_clean, x='GDP', y='Life Expectancy', 
    color=PURPLE, alpha=0.6, s=70, edgecolor=DARK
)
plt.title('GDP per Capita vs Life Expectancy', fontsize=14, color=DARK, fontweight='bold')
plt.xlabel('GDP per Capita ($)', fontsize=12)
plt.ylabel('Life Expectancy (Years)', fontsize=12)
plt.savefig('chart_2_gdp_life.png', dpi=300, bbox_inches='tight')
plt.show()

# WYKRES 3: GDP vs INFANT MORTALITY
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_clean, x='GDP', y='Infant Mortality', 
    color=MINT, alpha=0.8, s=70, edgecolor=DARK
)
plt.title('GDP per Capita vs Infant Mortality', fontsize=14, color=DARK, fontweight='bold')
plt.xlabel('GDP per Capita ($)', fontsize=12)
plt.ylabel('Infant Mortality (per 1000 births)', fontsize=12)
plt.savefig('chart_3_gdp_mortality.png', dpi=300, bbox_inches='tight')
plt.show()

# WYKRES 4: NUTRITION vs MORTALITY 
plt.figure(figsize=(10, 6))
sns.scatterplot(
    data=df_clean, x='Nutrition (Protein)', y='Infant Mortality', 
    color=PURPLE, alpha=0.6, s=70, edgecolor=MINT
)
plt.title('Nutrition (Protein Intake) vs Mortality', fontsize=14, color=DARK, fontweight='bold')
plt.xlabel('Daily Animal Protein Intake (Calories)', fontsize=12)
plt.ylabel('Infant Mortality (per 1000 births)', fontsize=12)
plt.savefig('chart_4_nutrition.png', dpi=300, bbox_inches='tight')
plt.show()
```
